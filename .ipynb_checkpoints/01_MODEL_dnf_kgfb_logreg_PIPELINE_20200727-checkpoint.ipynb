
## <font color ='green'> 4. A logisztikus regresszió modell illesztése és az előrejelzés létrehozása

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.datasets import make_classification
```

```python
# Create training and test sets
X, y = make_classification(weights=[0.1, 0.9], random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size= .3)
```

```python
# Fit the pipeline to the train set
pipeline.fit(X_train,y_train)
```

```python
# Predict the labels of the test set
y_pred = pipeline.predict(X_test)
```

```python
# Compute metrics
print(classification_report(y_test,y_pred))
```

```python
# a logreg modell elmentése helyi lemezre
import pickle
filename = 'logreg_20200718.sav'
pickle.dump(pipeline.named_steps['logreg'], open(filename, 'wb'))
```
